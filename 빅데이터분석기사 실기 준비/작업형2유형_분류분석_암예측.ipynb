# 데이터 나누기

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)
train_set, val_set = train_test_split(train_set, test_size=0.25, random_state=42)
```

```python
print(len(train_set))
print(len(val_set))
print(len(test_set))
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split=StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(data, data['diagnosis']):
    strat_train_set_=data.loc[train_index]
    strat_test_set=data.loc[test_index]
    
strat_train_set_=strat_train_set_.reset_index().drop('index', axis=1)
#split=StratifiedShuffleSplit(n_splits=1, test_size=0.25, random_state=42)
for train_index, val_index in split.split(strat_train_set_, strat_train_set_['diagnosis']):
    strat_train_set=strat_train_set_.loc[train_index]
    strat_val_set=strat_train_set_.loc[val_index]
```

```python
print(data['diagnosis'].value_counts() / len(data))
print()

print(strat_train_set['diagnosis'].value_counts() / len(strat_train_set))
print()

print(strat_val_set['diagnosis'].value_counts() / len(strat_val_set))
print()

print(strat_test_set['diagnosis'].value_counts() / len(strat_test_set))
print()
```

```python
print(len(data))
print(len(strat_train_set))
print(len(strat_val_set))
print(len(strat_test_set))
```

```python
X_train=train_set.drop('diagnosis', axis=1)
y_train=train_set['diagnosis']

X_val=val_set.drop('diagnosis', axis=1)
y_val=val_set['diagnosis']

X_test=test_set.drop('diagnosis', axis=1)
y_test=test_set['diagnosis']
```

# 모델 학습 - 확률적경사하강법

모델링 하기 전 결측값/이상치 처리 해줘야 하지만 이 데이터에는 딱히 할 필요 없을 것 같아 생략

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(loss='log', max_iter=500, random_state=42)
sgd_clf.fit(X_train, y_train)

print(sgd_clf.score(X_val, y_val))
print(sgd_clf.score(X_test, y_test))
```

# 모델 학습 - 로지스틱 회귀분석
다중클래스인 경우 소프트맥스 회귀

```python
# 로지스틱 회귀 할 때는 무조건 정규화!
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.fit_transform(X_val)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
print(log_reg.score(X_train, y_train))
print(log_reg.score(X_val, y_val))
print(log_reg.score(X_test, y_test))
```

```python
print(log_reg.coef_)
```

# 모델 학습 - 라쏘

```python
from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha=0.01)

lasso_model.fit(X_train, y_train)

print(lasso_model.score(X_train, y_train))
print(lasso_model.score(X_val, y_val))
print(lasso_model.score(X_test, y_test))
```

# 모델 학습 - 트리

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(max_depth=3)
tree_clf.fit(X_train, y_train)

print(tree_clf.score(X_train, y_train))
print(tree_clf.score(X_val, y_val))
print(tree_clf.score(X_test, y_test))
```

# 모델 학습 - 랜덤 포레스트

```python
from sklearn.ensemble import RandomForestClassifier
 
random_forest = RandomForestClassifier(n_estimators=50, max_leaf_nodes=16)
random_forest.fit(X_train, y_train)
 
print(random_forest.score(X_train, y_train))
print(random_forest.score(X_val, y_val))
print(random_forest.score(X_test, y_test))
```

# 모델 학습 - 배깅

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    max_samples=100, bootstrap=True, n_jobs=-1)
bag_clf.fit(X_train, y_train)

print(bag_clf.score(X_train, y_train))
print(bag_clf.score(X_val, y_val))
print(bag_clf.score(X_test, y_test))
```

# 모델 학습 - 부스팅 - 아다부스트

```python
from sklearn.ensemble import AdaBoostClassifier

ada_clf=AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=100,
    algorithm="SAMME.R", learning_rate=0.5)
ada_clf.fit(X_train, y_train)

print(ada_clf.score(X_train, y_train))
print(ada_clf.score(X_val, y_val))
print(ada_clf.score(X_test, y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier 

gb_clf = GradientBoostingClassifier(random_state=0)
gb_clf.fit(X_train, y_train)

print(gb_clf.score(X_train, y_train))
print(gb_clf.score(X_val, y_val))
print(gb_clf.score(X_test, y_test))
```

# 평가 및 모델 선택

```python
sgd_clf
log_reg
lasso_model
tree_clf
random_forest
bag_clf
ada_clf
gb_clf
```

```python
#정확도
from sklearn.metrics import accuracy_score

sgd_pred=sgd_clf.predict(X_test)
log_pred=log_reg.predict(X_test)
lasso_pred=lasso_model.predict(X_test)
tree_pred=tree_clf.predict(X_test)
rf_pred=random_forest.predict(X_test)
bag_pred=bag_clf.predict(X_test)
ada_pred=ada_clf.predict(X_test)
gb_pred=gb_clf.predict(X_test)
print(accuracy_score(y_test, sgd_pred))
print(accuracy_score(y_test, log_pred))
#print(accuracy_score(y_test, lasso_pred))
print(accuracy_score(y_test, tree_pred))
print(accuracy_score(y_test, rf_pred))
print(accuracy_score(y_test, bag_pred))
print(accuracy_score(y_test, ada_pred))
print(accuracy_score(y_test, gb_pred))
```

# 모델 세부 튜닝 

```python
#로지스틱
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100],
              'penalty': ['l1', 'l2']}

grid_search = GridSearchCV(LogisticRegression(), param_grid=param_grid, cv=5, n_jobs=-1, scoring="accuracy") 
grid_search.fit(X_train, y_train) 
print("best param : {}".format(grid_search.best_params_)) 
print("best acc : {}".format(grid_search.best_score_))
gb_pred=grid_search.predict(X_test)
print(accuracy_score(y_test, gb_pred))
```

```python
#결정트리
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion': ['gini', 'etropy'],
              'max_depth': range(1,10),
                'min_samples_split':range(1,10),
             'min_samples_leaf':range(1,5)}

grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid=param_grid, cv=5, n_jobs=-1, scoring="accuracy") 
grid_search.fit(X_train, y_train) 
print("best param : {}".format(grid_search.best_params_)) 
print("best acc : {}".format(grid_search.best_score_))
gb_pred=grid_search.predict(X_test)
print(accuracy_score(y_test, gb_pred))
```

```python
#랜덤포레스트
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion': ['gini', 'etropy'],
              'max_depth': range(1,10),
                'min_samples_split':range(1,10),
             'min_samples_leaf':range(1,5)}

grid_search = GridSearchCV(RandomForestClassifier(), param_grid=param_grid, cv=5, n_jobs=-1, scoring="accuracy") 
grid_search.fit(X_train, y_train) 
print("best param : {}".format(grid_search.best_params_)) 
print("best acc : {}".format(grid_search.best_score_))
gb_pred=grid_search.predict(X_test)
print(accuracy_score(y_test, gb_pred))
```
