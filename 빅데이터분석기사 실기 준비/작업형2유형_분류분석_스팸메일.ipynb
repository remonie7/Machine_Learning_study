# 샘플링

```python
data_log=pd.DataFrame(df_scale, columns=df_log.columns)
data_log['spam']=y
data_log.head(5)
```

```python
##필요하면 계층형 샘플링
from sklearn.model_selection import train_test_split
train_data, test_data=train_test_split(data_log, test_size=0.2, random_state=42)
train_data, val_data=train_test_split(train_data, test_size=0.25, random_state=42)
```

```python
print(len(train_data))
print(len(val_data))
print(len(test_data))
```

```python
x_train=train_data.drop('spam', axis=1)
y_train=train_data['spam']

x_test=test_data.drop('spam', axis=1)
y_test=test_data['spam']

x_val=val_data.drop('spam', axis=1)
y_val=val_data['spam']
```

```python
data_log['spam'].value_counts() ## 
```

# 모델링

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score

glm = LogisticRegression()
glm.fit(x_train, y_train)
print(glm.score(x_train, y_train))
print(glm.score(x_val, y_val))
print(glm.score(x_test, y_test))
print(roc_auc_score(y_val, glm.predict(x_val)))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100],
              'penalty': ['l1', 'l2']}

grid_search = GridSearchCV(LogisticRegression(), param_grid=param_grid, cv=10, n_jobs=-1, scoring="accuracy") 
grid_search.fit(x_train, y_train) 
print("best param : {}".format(grid_search.best_params_)) 
print("best acc : {}".format(grid_search.best_score_))
gb_pred=grid_search.predict(x_val)
print(roc_auc_score(y_val, gb_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, roc_auc_score

tree=DecisionTreeClassifier()
tree.fit(x_train, y_train)
print(tree.score(x_train, y_train))
print(tree.score(x_val, y_val))
print(tree.score(x_test, y_test))
print(roc_auc_score(y_val, tree.predict(x_val)))
```
