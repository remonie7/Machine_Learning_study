## 분석 기법 적용
#### 회귀 분석  // 로버스트회귀분석 // LOESS

- 모형의 설명력 : 결정계수 R^2 (= ssr/sst = (1-sse)/sst )

- 계수의 유의성 : 왈드 검정 (p-value)
- 모형의 설명력 : 결정계수, AIC
