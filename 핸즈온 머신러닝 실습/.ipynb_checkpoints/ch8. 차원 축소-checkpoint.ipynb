## 차원 축소를 위한 접근 방법

#### 투영
대부분의 실전 문제에서, 모든 훈련 샘플이 사실 고차원 공간 안의 저차원 부분 공간에 가까이 놓여 있는 경우가 많다.

이런 경우 훈련 샘플들을 저차원 부분 공간에 수직으로 투영하여 저차원 데이터셋 얻을 수 있다.


그러나 스위스 롤 데이터처럼 부분 공간이 뒤틀리거나 휘어 있는 경우도 있으므로, (이 경우 스위스 롤을 펼쳐서 저차원 데이터셋 얻어야 함) 그러므로 투영이 항상 좋은 차원 축소의 방법은 될 수 없다

#### 매니폴드 학습

- 예를 들어, 2D 매니폴드는 고차원 공간에서 휘어지거나 뒤틀린 2D 모양이다 (스위스롤은 2D 매니폴드의 한 예)
- 매니폴드 학습 : 훈련 샘플이 놓여 있는 매니폴드를 모델링
- 매니폴드 가정(가설) : 실제 고차원 데이터셋이 더 낮은 저차원 매니폴드에 가깝게 놓여 있다는 가정
- 매니폴드 가정이 늘 좋은 건 아니다. 매니폴드를 펼치기 전에는 결졍 경계가 간단하나, 펼치고 난 후에는 결정 경계가 더 복잡해지는 경우의 데이터도 존재한다.

## PCA
- 주성분분석(PCA) : 데이터에 가장 가까운 초평면을 정의한 다음, 데이터를 이 평면에 투영시키는 차원 축소 알고리즘

#### 분산 보존
투영하였을 때 분산이 최대로 보존되는 초평면을 선택해야 한다!

#### 주성분
- 분산이 최대인 축을 찾는다
- 첫번째 축에 직교하고 남은 분산을 최대한 보존하는 두 번째 축을 찾는다
- 첫번째, 두번째 축에 직교하면서 남은 분산을 최대화하는 세 번째 축을 찾는다
- 반복


- i번째 축을 정의하는 단위 벡터를 i번째 주성분이라고 한다.



- 훈련 세트의 주성분 찾는 방법 : 특잇값 분해(SVD) 라는 표준 행렬 분해 기술으로, 훈련 세트 행렬 X를 세 개 행렬의 점곱인 U * SIGMA * V^T 로 분하핼 수 있다. 여기서 찾고자 하는 주성분이 V에 담겨 있다.

- V = ( c_1벡터  c_2벡터  ... c_n벡터 )

```python
import numpy as np


## 데이터셋 만들기
np.random.seed(4)
m = 60
w1, w2 = 0.1, 0.3
noise = 0.1

angles = np.random.rand(m) * 3 * np.pi / 2 - 0.5
X = np.empty((m, 3))
X[:, 0] = np.cos(angles) + np.sin(angles)/2 + noise * np.random.randn(m) / 2
X[:, 1] = np.sin(angles) * 0.7 + noise * np.random.randn(m) / 2
X[:, 2] = X[:, 0] * w1 + X[:, 1] * w2 + noise * np.random.randn(m)


X
```

```python
X_centered = X - X.mean(axis=0)
U, s, Vt = np.linalg.svd(X_centered)
c1 = Vt.T[:,0]
c2 = Vt.T[:,1]
Vt.T  # V

### X 데이터가 3차원이니까 c_3까지 나오는 것 (모든 주성분 추출)
```

#### d차원으로 투영하기
초평면에 훈련 세트를 투영하기 위해서는, 행렬 X와, 첫 d개의 주성분을 담은 (V의 첫 d열로 구성된) 행렬 W_d를 점곱한다
- X * W_d

```python
W2 = Vt.T[:, :2]
X2D = X_centered.dot(W2)
X2D # 3차원이 2차원으로 줄어들었다!
```

#### 사이킷런 이용하기
사이킷런의 PCA 모델은 SVD 분해 방법을 사용하여 구현

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X2D = pca.fit_transform(X)

pca.components_.T  # 주성분벡터 V
```

#### 설명된 분산의 비율
- explained_variance_ratio_ 변수에 저장되어 있다.
- 각 주성분의 축을 따라 있는 데이터셋의 분산 비율 나타냄

```python
pca.explained_variance_ratio_

# 데이터셋 분산의 84.2%가 첫 번째 축에 놓여 있음
# 데이터셋 분산의 14.6%가 두 번째 축에 놓여 있음
# 세 번째 축에는 1.2% 미만이 남아 있을 것이므로 아주 작은 정보가 들어있다고 생각하면 됨
```

#### 적절한 차원 수 선택하기
- 충분한 분산(95%정도)이 될 때까지 더해야 할 차원 수 선택하는 방법이 일반적
- 시각화를 위한 차원 축소의 경우에는 차원을 2개나 3개 정도로 줄이기도 함

```python
from sklearn.datasets import fetch_openml 
mnist = fetch_openml('mnist_784', version=1, cache=True) 
X, y = mnist["data"], mnist["target"]
y = np.array(list(map(int, y)))

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
pca = PCA()
pca.fit(X_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d=np.argmax(cumsum>=0.95)+1 

#필요 차원 수 계산하고 난 뒤 차원수 설정하여 PCA 다시 실행
#pca=PCA(n_components=d)
pca=PCA(n_components=0.95) # d 구하여 실행하는 것보다 이 방법 더 선호

X_reduced = pca.fit_transform(X_train)
```

#### 압축을 위한 PCA
압축된 데이터셋에 PCA 투영의 변환을 반대로 적용하여 원본 차원으로 되돌릴 수도 있다.
- inverse_transform() 메서드

```python
pca = PCA(n_components=154)
X_reduced = pca.fit_transform(X_train)
X_recovered = pca.inverse_transform(X_reduced)
```

#### 점진적 PCA
- PCA 구현의 문제는 SVD 알고리즘을 실행하기 위해 전체 훈련 세트를 메모리에 올려야 한다는 것
- 이 단점을 보완하는 방법 : 점진적 PCA(IPCA)
- IPCA : 훈련 세트를 미니배치로 나눈 뒤 IPCA 알고리즘에 한 번에 하나씩 주입

```python
from sklearn.decomposition import IncrementalPCA

n_batches = 100
inc_pca = IncrementalPCA(n_components=154)
for X_batch in np.array_split(X_train, n_batches):
    inc_pca.partial_fit(X_batch)      # fit()이 아닌 partial_fit()

X_reduced = inc_pca.transform(X_train)
```

#### 랜덤 PCA
- 확률적인 알고리즘으로, 첫 d개의 주성분에 대한 근삿값을 빠르게 찾는다
- d가 n보다 많이 작으면 매우 빨라짐

```python
rnd_pca = PCA(n_components=154, svd_solver="randomized")
X_reduced = rnd_pca.fit_transform(X_train)
```

## 커널 PCA (kPCA)
- 커널 PCA : 차원 축소를 위한 복잡한 비선형 투영
- 이 기법은 투영된 후에 샘플의 군집을 유지하거나 꼬인 매니폴드에 가까운 데이터셋 펼칠 때도 유용

```python
#데이터 만들기
from sklearn.datasets import make_swiss_roll
X, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)

#커널 pca
from sklearn.decomposition import KernelPCA
rbf_pca = KernelPCA(n_components=2, kernel="rbf", gamma=0.04)
X_reduced = rbf_pca.fit_transform(X)
```

```python
#선형 커널(단순pca), rbf커널, 시그모이드(로지스틱)커널 사용해 2차원으로 축소시킨 스위스 롤의 모습 
from sklearn.decomposition import KernelPCA

lin_pca = KernelPCA(n_components = 2, kernel="linear", fit_inverse_transform=True)
rbf_pca = KernelPCA(n_components = 2, kernel="rbf", gamma=0.0433, fit_inverse_transform=True)
sig_pca = KernelPCA(n_components = 2, kernel="sigmoid", gamma=0.001, coef0=1, fit_inverse_transform=True)

y = t > 6.9

plt.figure(figsize=(11, 4))
for subplot, pca, title in ((131, lin_pca, "선형 커널"), (132, rbf_pca, "RBF 커널, $\gamma=0.04$"), (133, sig_pca, "시그모이드 커널, $\gamma=10^{-3}, r=1$")):
    X_reduced = pca.fit_transform(X)
    if subplot == 132:
        X_reduced_rbf = X_reduced
    
    plt.subplot(subplot)
    #plt.plot(X_reduced[y, 0], X_reduced[y, 1], "gs")
    #plt.plot(X_reduced[~y, 0], X_reduced[~y, 1], "y^")
    plt.title(title, fontsize=14)
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=t, cmap=plt.cm.hot)
    plt.xlabel("$z_1$", fontsize=18)
    if subplot == 131:
        plt.ylabel("$z_2$", fontsize=18, rotation=0)
    plt.grid(True)

plt.show()
```

#### 커널 선택과 하이퍼파라미터 튜닝
- kPCA는 비지도 학습이기 때문에 좋은 커널과 하이퍼파라미터를 선택하기 위한 명확한 성능 측정 기준은 없지만, 차원 축소는 종종 지도 학습의 전처리 단계로 활용되므로 그리드 탐색을 사용하여 주어진 문제에서 성능이 가장 좋은 커널과 하이퍼파라미터를 선택할 수 있다.



```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

clf = Pipeline([
        ("kpca", KernelPCA(n_components=2)),
        ("log_reg", LogisticRegression(solver='liblinear'))
    ])

param_grid = [{
        "kpca__gamma": np.linspace(0.03, 0.05, 10),
        "kpca__kernel": ["rbf", "sigmoid"]
    }]

grid_search = GridSearchCV(clf, param_grid, cv=3)
grid_search.fit(X, y)
```

```python
print(grid_search.best_params_)
```

- 가장 낮은 재구성 오차를 만드는 커널과 하이퍼파라미터 선택하는 방식도 있음
- 재구성 원상을 얻게 되면 원본 샘플과의 제곱 거리를 특정할 수 있고, 재구성 원상의 오차를 최소화하는 커널과 하이퍼파라미터 선택할 수 있음


- 재구성 원상 오차를 최소화하는 커널과 하이퍼파라미터를 찾기 위해 교차 검증으로 그리드 탐색 사용 가능ㅁ

```python
#재구성
rbf_pca = KernelPCA(n_components=2, kernel="rbf", gamma=0.0433, fit_inverse_transform=True)
X_reduced = rbf_pca.fit_transform(X)
X_preimage = rbf_pca.inverse_transform(X_reduced)

#재구성 오차 계산
from sklearn.metrics import mean_squared_error
mean_squared_error(X, X_preimage)
```

## LLE (지역 선형 임베딩)
- 비선형 차원 축소(NLDR) 기술
- 투영에 의존하지 않는 매니폴드 학습



- LLD는 먼저 각 훈련 샘플이 가장 가까운 이웃에 얼마나 선형적으로 연관되어 있는지 측정
- 그런 다음 국부적인 관계가 가장 잘 보존되는 훈련 세트의 저차원 표현을 찾는다
- 이는 특히 잡음이 너무 많지 않은 경우 꼬인 매니폴드를 펼치는 데 잘 작동한다

```python
from sklearn.manifold import LocallyLinearEmbedding

lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_reduced = lle.fit_transform(X)



plt.title("LLE를 사용하여 펼쳐진 스위스 롤", fontsize=14)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=t, cmap=plt.cm.hot)
plt.xlabel("$z_1$", fontsize=18)
plt.ylabel("$z_2$", fontsize=18)
plt.axis([-0.065, 0.055, -0.1, 0.12])
plt.grid(True)

plt.show()
```
